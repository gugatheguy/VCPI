### Unsharp mask

$I = I + \alpha(I - G)$

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/IMG_20211023_165609.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
fig = plt.figure(figsize=(30,10))
plt.imshow(img, cmap='gray')
```

```python
gaussian = cv2.GaussianBlur(img, (45, 45), 0)
unsharp_mask = cv2.addWeighted(img, 1.0, gaussian, -1.0, 0)

fig = plt.figure(figsize=(30,10))
plt.imshow(unsharp_mask, cmap='gray')
```

```python
sharped = cv2.addWeighted(img, 2.0, gaussian, -1.0, 0)

fig = plt.figure(figsize=(30,10))
plt.imshow(sharped, cmap='gray')
```

Note: since cv2 caps the result of addWeighted to [0,255] if we opted to simply add the result
