```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img = cv2.imread('images/lena_salt_and_pepper.png')
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### Compare the various filters to remove the noise in the image above

```python
resinc = cv2.medianBlur(img, 3)
res = cv2.bilateralFilter(resinc, 9, 75, 75)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
res = cv2.medianBlur(img, 3)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
res = cv2.GaussianBlur(img, (13,13), -1)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
res = cv2.bilateralFilter(img, 50, 75, 75)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
res = cv2.fastNlMeansDenoising(img,None,10,7,21)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```
