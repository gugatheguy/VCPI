```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

### Median Blur

This is a non-linear filter that can't be applied as a convolution.

For each pixel, the median intensity of its neighbourhood becomes its value

Another alternative is **Conservative Smoothing**

- For each pixel

    - compute intensity max and min of neighbourhood
    - if pixel intensity > max
        - pixel intensity = max
    - else if pixel intensity < min
        - pixel intensity = min
    



### Bilateral Filter

$BF(I)_p = \frac{1}{W_p}  \sum_{q \in S} G_{\sigma_s}(||p-q||) G_{\sigma_r}(||I_p - I_q)$

The first Guassian is the same as in the Gaussian filter. The second provides smoothing based on the intensity difference between the intensity of pixel itself and the intensities of the pixels covered by the kernel.

Setting $\sigma_r$ to infinity (some very large value) resolves the above function to a Gaussian filter.

$W_p$ is the normalization factor to ensure that the image remains with the same intensity average and is equal to:

$W_p = \sum_{q \in S} G_{\sigma_s}(||p-q||) G_{\sigma_r}(||I_p - I_q)$



```python
img2 = cv2.imread('images/lena.jpg')
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
```

```python
res = cv2.medianBlur(img2, 9)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python
res = cv2.GaussianBlur(img2, (7,7), -1)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```

```python

res = cv2.bilateralFilter(img2, 9, 75, 75)
fig = plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
```
